# ImageStitch and Aerial Photos
Aerial photos are usually taken with overlaps along the flight lines, it is a perfect application for ImageStitch to generate a mosaic of aerial photos.
More on [Concept of Aerial Photography](https://www.nrcan.gc.ca/maps-tools-publications/satellite-imagery-air-photos/air-photos/national-air-photo-library/about-aerial-photography/concepts-aerial-photography/9687)

## 1. Get the aerial photos
We use the same data set used in the previous ImageStitch notebook.  
[FlightLine 3196](https://data.cresis.ku.edu/data/photogrammetry/Out_FlightLine_3196/), it is somewhere in Antarctica.

```
(* extract image names *)
datalocation = "https://data.cresis.ku.edu/data/photogrammetry/Out_FlightLine_3196/";
rawdata = Import[datalocation,"Data"];
imagelinks = Select[rawdata[[All,1]], StringMatchQ["*CA3196*"]];
imagelinks
```

```
(* import and resize image to 1/4 *)
images = ImageResize[Import[datalocation <> #],Scaled[1/4]]& /@ imagelinks
```

## 2. ImageStitch 
Update ImageStitch function in 13.2

```
ImageStitch[images]
```
